```python
# y = wx +b
# y ise label
y = dataFrame["Fiyat"].values

#x ise feature yani zellik demektir
x = dataFrame[["BisikletOzellik1","BisikletOzellik2"]].values

#values kullanmaz ise pandas series
#values kullanr isen numpy dizisi
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.33,random_state=15)
```

```python
#burada y iin degil ve ozellikler icin yap, asl onlar on planda
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
#kucultmek istedigin veriyi sec
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
#model oluturma
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential #katmanlar ile almay salar
```

```python
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(4,activation="relu"))#hidden layer ekleme
model.add(Dense(4,activation="relu"))#hidden layer ekleme
model.add(Dense(4,activation="relu"))#hidden layer ekleme
```

```python
#kt katman ekleme
model.add(Dense(1))
```

```python
model.compile(optimizer = "rmsprop",loss="mse")#her eyi tamamla ve optimizer ile istediini se fonksiyondan ve loss ile kayp fonksiyonu ekl
```

```python
#del model
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
loss = model.history.history["loss"]
```

```python
sbn.lineplot(x=range(len(loss)),y=loss)
```

```python
trainLoss = model.evaluate(x_train,y_train, verbose=0)
```

```python
testLoss = model.evaluate(x_test,y_test,verbose=0)
```
