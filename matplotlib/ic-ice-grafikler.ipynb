```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numpyDizisi1 = np.linspace(0,10,20) #bellir aralıklar ile liste oluştur
numpyDizisi1
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.get_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```
