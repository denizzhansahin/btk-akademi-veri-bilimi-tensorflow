```python
#bir sıra iki kolon
plt.subplot(2,2,1) #alt grafik #yan yana olup olmadığı, kaçtane ve hangisi
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(2,2,2) #1 sıra 2 kolon 2.grafik
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
plt.subplot(2,2,3) #alt grafik #yan yana olup olmadığı, kaçtane ve hangisi
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(2,2,4) #1 sıra 2 kolon 2.grafik
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```
