```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

numpyDizisi1 = np.linspace(0,10,20) #bellir aralıklar ile liste oluştur
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
yeniFigur = plt.figure()
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi2,label="numoyDizisi ** 2")
yeniEksen.legend()
```

```python
yeniFigur = plt.figure()
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label="numoyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpyDizisi ** 3")
yeniEksen.legend()
```
