```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
numpyDizisi2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"g")
benimEksen.plot(numpyDizisi2,numpyDizisi1,"b")
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,color="#3A95A8")#hex code
benimEksen.plot(numpyDizisi2,numpyDizisi1,color="yellow")#renk adı
```
