```python
import numpy as np
```

```python
import pandas as pd
```

```python
sozluk1 =  {"Isim":["Ahmet","Mehmet","Zeynep","Deniz"],
            "Spor":["Koşu","Yüzme","Koşu","Basketbol"],
            "Kalori":[100,200,300,400]
            }
```

```python
dataFrame1 = pd.DataFrame(sozluk1,index=[0,1,2,3])
```

```python
dataFrame1
```

```python
sozluk2 =  {"Isim":["Osman","Levent","Atlas","Fatma"],
            "Spor":["Koşu","Yüzme","Koşu","Basketbol"],
            "Kalori":[200,100,50,400]
            }

dataFrame2 = pd.DataFrame(sozluk2,index=[4,5,6,7])
```

```python
dataFrame2
```

```python
sozluk3 =  {"Isim":["Ayşe","Mahmut","Duygu","Nur"],
            "Spor":["Koşu","Yüzme","Badminton","Tenis"],
            "Kalori":[300,400,500,250]
            }

dataFrame3 = pd.DataFrame(sozluk3,index=[8,9,10,11])
```

```python
dataFrame3
```

```python
#concatenation - birleştirme
```

```python
pd.concat([dataFrame1,dataFrame2,dataFrame3])
```

```python
pd.concat([dataFrame1,dataFrame2,dataFrame3],axis=1)
```

```python
pd.concat([dataFrame1,dataFrame2,dataFrame3],axis=0)
```

```python
#merge - birleştirme - merging - kaynaştırmak
```

```python
mergeSozluk1 = {"Isim":["Ahmet","Mehmet","Zeynep","Deniz"],
               "Spor":["Koşu","Yüzme","Koşu","Basketbol"]
               }
```

```python
mergeDataFrame1 = pd.DataFrame(mergeSozluk1)
```

```python
mergeDataFrame1
```

```python
mergeSozluk2 = {"Isim":["Ahmet","Mehmet","Zeynep","Deniz"],
               "Kalori":[100,200,150,250]
               }
```

```python
mergeDataFrame2 = pd.DataFrame(mergeSozluk2)
```

```python
mergeDataFrame2
```

```python
pd.merge(mergeDataFrame1,mergeDataFrame2,on="Isim")
```
